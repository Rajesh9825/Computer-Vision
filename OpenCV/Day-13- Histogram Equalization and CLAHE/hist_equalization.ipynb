```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("./giraffe-Kenya.png")

if image is None:
    print("Issue with image loading.")
else:
    print("Image loaded successfully.")
```

```python
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image_greyscale = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
hist , bins = np.histogram(image_greyscale.flatten(),256,[0,256])

# Perform Histogram Equalization

equalized_image = cv2.equalizeHist(image_greyscale) 

equalized_hist , bins = np.histogram(equalized_image.flatten(),256,[0,256])

# Plot the results
plt.figure(figsize=(12,6))
plt.subplot(2,2,1)
plt.imshow(image_greyscale,cmap="gray")
plt.title("Original Image")

plt.subplot(2,2,2)
plt.plot(hist,color="blue")
plt.title("Original Image Histogram")

plt.subplot(2,2,3)
plt.imshow(equalized_image,cmap="gray")
plt.title("Equalized Image")

plt.subplot(2,2,4)
plt.plot(equalized_hist,color="green")
plt.title("Equalized Image Histogram")


plt.tight_layout()
plt.show()
```

```python
#LAB
# YUV used in television broadcast or media player,
# y is luminance and uv are chrominance
```

```python
image_yuv = cv2.cvtColor(image,cv2.COLOR_BGR2YUV)
image_yuv[:,:,0] = cv2.equalizeHist(image_yuv[:,:,0]) # y channel only

equalizeded_color_image = cv2.cvtColor(image_yuv,cv2.COLOR_YUV2BGR)
equalizeded_color_image_RGB = cv2.cvtColor(equalizeded_color_image,cv2.COLOR_BGR2RGB)


# Plot the results
plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.imshow(image_rgb)
plt.title("Original Image")

plt.subplot(1,2,2)
plt.imshow(equalizeded_color_image_RGB)
plt.title("Equalized RGB Image")


plt.tight_layout()
plt.show()
```
