```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("./giraffe-Kenya.png")
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
image_grey = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

equalized_image = cv2.equalizeHist(image_grey)

clahe = cv2.createCLAHE(clipLimit=2.0,tileGridSize=(4,4))

clahe_equalized_image = clahe.apply(image_grey)

plt.figure(figsize=(12,8))

plt.subplot(1,3,1)
plt.imshow(image_grey,cmap="gray")
plt.title("Original Image")


plt.subplot(1,3,2)
plt.imshow(equalized_image,cmap="gray")
plt.title("Equalized Image")

plt.subplot(1,3,3)
plt.imshow(clahe_equalized_image,cmap="gray")
plt.title("Clahe Equalized Image")

plt.show()
```

```python
yuv_image = cv2.cvtColor(image,cv2.COLOR_BGR2YUV)

yuv_image[:,:,0] = clahe.apply(yuv_image[:,:,0])

yuv_image_BGR = cv2.cvtColor(yuv_image,cv2.COLOR_YUV2BGR)

yuv_image_rgb = cv2.cvtColor(yuv_image_BGR,cv2.COLOR_BGR2RGB)


plt.figure(figsize=(12,8))

plt.subplot(1,2,1)
plt.imshow(image_rgb)
plt.title("Original RGB Image")


plt.subplot(1,2,2)
plt.imshow(yuv_image_rgb)
plt.title("Clahe Equalized RGB Image")


plt.show()
```
