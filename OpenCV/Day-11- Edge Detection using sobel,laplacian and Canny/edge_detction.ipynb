```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("./giraffe-Kenya.png",cv2.IMREAD_GRAYSCALE)
```

```python
# Sobel Edge Detection
sobel_X = cv2.Sobel(image,cv2.CV_64F,1,0,ksize=3)
sobel_Y = cv2.Sobel(image,cv2.CV_64F,0,1,ksize=3)


sobel_combined = cv2.magnitude(sobel_X,sobel_Y)


# Laplacian 
laplacian = cv2.Laplacian(image,cv2.CV_64F,ksize=3) # it gives +ve and -ve values
laplacian_abs = cv2.convertScaleAbs(laplacian) # to avoid negative we use absolute


# Canny 
canny_image = cv2.Canny(image,180,200)


titles = ["Original","Sobel","Laplacian_All","Laplacian_Abs","Canny"]
images = [image,sobel_combined,laplacian,laplacian_abs,canny_image]

plt.figure(figsize=(20,16))
for i in range(len(titles)):
    plt.subplot(3,2,i+1)
    plt.imshow(images[i],cmap='grey')
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.show()



```
