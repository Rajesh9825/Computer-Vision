```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("./giraffe-Kenya.png")
```

```python
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image_hsv = cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
plt.imshow(image_rgb)
```

```python
# Color Thresholding
lower_white = np.array([0,0,190]) # in terms of HSV, keeping brightness V=190, H = 0 degree
upper_white = np.array([150,50,255]) # array in terms of HSV, keeping H as 30 degree because we are not targeting pure white it has some brown and orange into the color.
```

```python
mask = cv2.inRange(image_hsv,lower_white,upper_white)

plt.imshow(mask) # this mask is not black and white because matplotlib give in terms of purple and yellow
```

```python
black_hsv = np.uint8([[[120,255,0]]]) # choose the which colr want to raplace in hsv then convert into BGR
black_bgr = cv2.cvtColor(black_hsv,cv2.COLOR_HSV2BGR)[0][0] 
black_hsv
```

```python
image_result = image.copy()

image_result[mask>0] = black_bgr
```

```python
image_result_rgb = cv2.cvtColor(image_result,cv2.COLOR_BGR2RGB)
```

```python
# Plot 
fig , axs = plt.subplots(1,3,figsize=(12,6))

axs[0].imshow(image_rgb)
axs[0].set_title("original_image")
axs[0].axis('off')

axs[1].imshow(mask)
axs[1].set_title("Binary Mask")
axs[1].axis('off')

axs[2].imshow(image_result_rgb)
axs[2].set_title("Output_image")
axs[2].axis('off')

plt.tight_layout()
plt.show()
```
