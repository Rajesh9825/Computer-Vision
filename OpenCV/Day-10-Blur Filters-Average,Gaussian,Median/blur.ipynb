```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("./giraffe-Kenya.png")
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
# Average Blur
kernel_s = 5
average_blur = cv2.blur(image_gray,(kernel_s,kernel_s)) # bigger the kernel size then more the blurriness effect is going to be.
plt.imshow(average_blur,cmap='gray')
```

```python
# Gaussian Blur
kernel_s = 11 # it should be odd number in this case.

gaussian_blur = cv2.GaussianBlur(image_gray,(kernel_s,kernel_s),0)
# sigmaX is standard deviation of gaussian in the X direction, the higher the X means it is more smoother.
plt.imshow(gaussian_blur,cmap='gray')
```

```python
# Median Blur

median_blur = cv2.medianBlur(image_gray,kernel_s)

plt.imshow(median_blur,cmap='gray')
```

```python
plt.figure(figsize=(14,12))

plt.subplot(2,2,1)
plt.title("Originial Gray Image")
plt.imshow(image_gray,cmap="gray")
plt.axis("off")

plt.subplot(2,2,2)
plt.title("Average Blur Image")
plt.imshow(average_blur,cmap="gray")
plt.axis("off")

plt.subplot(2,2,3)
plt.title("Gaussian Blur Image")
plt.imshow(gaussian_blur,cmap="gray")
plt.axis("off")

plt.subplot(2,2,4)
plt.title("Median Blur Image")
plt.imshow(median_blur,cmap="gray")
plt.axis("off")

plt.tight_layout()
plt.show()

# You should either go with average blur or gaussian blur when you want to smoothen your images,
# and in terms of noise you should go with median blur.
```
